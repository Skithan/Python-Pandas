**Example code**  The following code snippet uses Python graphics library Turtle. Execute the code. There will be a pop-up window where the image of a bar chart will be rendered.

More info about Turtle library is here: https://docs.python.org/3.8/library/turtle.html

```python
import turtle
turtle.clearscreen()
t = turtle.Turtle()
scrn = turtle.Screen()
scrn.bgcolor("white")
t.speed(10)


def drawBar(t, height, label):
    """ Get turtle t to draw one bar, of height. """
    t.begin_fill()    # start filling this shape
    t.right(90)
    t.penup()
    t.forward(2)
    t.left(90)
    t.write(str(label))
    t.left(90)
    t.forward(2)
    t.pendown()
    
    t.forward(height)
    t.write(str(height))
    t.right(90)
    t.forward(40)
    t.right(90)
    t.forward(height)
    t.left(90)
    t.end_fill()                 # stop filling this shape
    
    

k = input ("Please input k: ")
numK = int (k)
if numK < 3 or numK > 10:
    print ("Out of range, please specify a value between 3 and 10.")
    
else:
    print ("We will show longest ", numK , " words present in the text") 


    
# write your code below
    passage = "I went to the woods because I wished to live deliberately, to front only the essential facts of life, and see if I could not learn what it had to teach, and not, when I came to die, discover that I had not lived. I did not wish to live what was not life, living is so dear; nor did I wish to practice resignation, unless it was quite necessary. I wanted to live deep and suck out all the marrow of life, to live so sturdily and Spartan-like as to put to rout all that was not life, to cut a broad swath and shave close, to drive life into a corner, and reduce it to its lowest terms, and, if it proved to be mean, why then to get the whole and genuine meanness of it, and publish its meanness to the world; or if it were sublime, to know it by experience, and be able to give a true account of it in my next excursion."
    passage = passage.replace('.',' ').replace('-',' ').replace(',',' ')
    passage = passage.lower()

    passage = passage.split()
    passage = list(dict.fromkeys(passage))

    sortedPassage = sorted(passage, key=len, reverse=True)
    
    xs = []
    xlabels = []

    for i in range(numK):
        xs.append(len(sortedPassage[i]))
        xlabels.append(sortedPassage[i])

border = 9
wn = turtle.Screen()             # Set up the window and its attributes
wn.bgcolor("white")

tess = turtle.Turtle()           # create tess and set some attributes
tess.pensize(3)


# Here we specify the data to use for rendering the bar chart
# hardcoded for 10 bars: don't do this in your progeam for Q3 



# xs = [4, 17, 20, 6, 3, 5, 4, 1, 7, 15]  # here is the data for the height of the bars
# xlabels = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j"]  # here is the data for the text labels of the bars
maxheight = max(xs)
numbars = len(xs)

wn.setworldcoordinates(0-border, 0-border, 40*numbars+border, maxheight+border)
tess.color("blue")
tess.fillcolor("lightgreen")


kvp = zip(xs, xlabels)
for (ht, lbl) in kvp:
    drawBar(tess, ht, lbl)

print("DJD")
    
# close resources properly 
# https://stackoverflow.com/questions/7217405/turtle-graphics-not-responding
t.hideturtle()
turtle.done() 
try:
    turtle.bye()   
except turtle.Terminator:
    pass
```
