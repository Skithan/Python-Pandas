**Q2** Load the file "movies.csv" into a pandas dataframe. This file contains information, specifically its movieId,  about many movies.


**Q2.1** Combine this dataframe with the dataframe you created in **Q1** such that the new dataframe will contain information about all the Oscar winning movies regardless of whether they have valid movieIds.

**Q2.2** In the newly created dataframe, if any row that has a movieId value NaN, replace it with -1. For those rows, if the title is NaN, replace it with "", and if the releasedYear is NaN, replace it with 0.  

```python
import pandas as pd
# write your code below.
df2 = pd.read_csv("movies.csv", delimiter=';')

#Q2.1
df3=pd.merge(df1,df2, how='left', left_on=['name','released_year'], right_on=['title','releasedYear'])


#Q2.2
df3['movieId'] = df3['movieId'].fillna(-1)
df3['title'] = df3['title'].fillna("")
df3['releasedYear'] = df3['releasedYear'].fillna(0)

df3['movieId'] = df3['movieId'].astype(int)
df3['releasedYear'] = df3['releasedYear'].astype(int)

df3



```
